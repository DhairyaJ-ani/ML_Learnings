```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining the positive and negative classes
positive_points = np.array([(-2,3), (0,1), (2,-1)])
negative_points = np.array([(-2,1), (0,-1), (2,-3)])
```

```python
#Plot initial data
plt.figure(figsize=(8,6))
plt.scatter(positive_points[:,0], positive_points[:,1], color='blue', marker='o', label='Positive Class')
plt.scatter(negative_points[:,0], negative_points[:,1], color='red', marker='x', label='Negative Class')
plt.legend(loc='best')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Initial Data points')
plt.grid(True)
plt.show()
```

```python
X = np.vstack((positive_points, negative_points))
Y= np.vstack([1]*len(positive_points) + [-1]*len(negative_points))
```

```python
def perceptron(x, y, steps=10):
    theta = np.zeros(x.shape[1])
    theta0 = 0
    for t in range(steps):
        for i, xi in enumerate(x):
            if y[i] * (np.dot(xi, theta) + theta0) <= 0:
                theta += y[i] * xi
                theta0 += y[i]
                break
        plot_decision_boundary(theta, theta0, t+1)
    return theta, theta0
```

```python
def plot_decision_boundary(theta, theta0, iterations):
    plt.figure(figsize=(8,6))
    plt.scatter(positive_points[:,0], positive_points[:,1], color='blue', marker='o', label='Positive Class')
    plt.scatter(negative_points[:,0], negative_points[:,1], color='red', marker='x', label='Negative Class')

    x_values = np.linspace(-3,3,2)

    if theta[1] != 0:
        y_values = -(theta[0]* x_values + theta0)/theta[1]
        plt.plot(x_values, y_values, color='green', label= f'Iteration {iterations}')
    else:
        x_values = -theta0/theta[0]
        plt.axvline(x_values, color='green', label= f'Iteration {iterations}')

    plt.xlim(-3,3)
    plt.ylim(-4,4)
    plt.legend(loc='best')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(f'Iteration {iterations}')
    plt.grid(True)
    plt.show()
```

```python
theta_final, theta0_final = perceptron(X, Y, steps=10)
```
